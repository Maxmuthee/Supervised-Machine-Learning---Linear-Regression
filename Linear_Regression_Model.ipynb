```python
#Linear Model - Ordinary Least Squares (OLS)
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
myData = load_diabetes()
X_train, X_test, y_train, y_test= train_test_split(myData.data, myData.target, random_state = 0)# you can ignore the random state unless asked otherwise
LR = LinearRegression().fit(X_train, y_train)#.fit() is for training
print("Training Accuracy: ", LR.score(X_train, y_train))
print("Testing Accuracy: ", LR.score(X_test, y_test))
```

```python
# Linear Model - Ridge Regression - Picks line with the least gradient
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
myData = load_diabetes()
X_train, X_test, y_train, y_test= train_test_split(myData.data, myData.target, random_state = 0)# you can ignore the random state unless asked otherwise
LR = Ridge(alpha = 0.01).fit(X_train, y_train)#.fit() is for training
print("Training Accuracy: ", LR.score(X_train, y_train))
print("Testing Accuracy: ", LR.score(X_test, y_test))
```

```python
# Linear Model -Lasso Regression
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
myData = load_diabetes()
X_train, X_test, y_train, y_test= train_test_split(myData.data, myData.target, random_state = 0)# you can ignore the random state unless asked otherwise
LR = Lasso(alpha = 0.5).fit(X_train, y_train)#.fit() is for training
print("Training Accuracy: ", LR.score(X_train, y_train))
print("Testing Accuracy: ", LR.score(X_test, y_test))
print("Number of Features", sum(LR.coef_ != 0))
```
